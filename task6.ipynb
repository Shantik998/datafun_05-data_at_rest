## Creating Histogram

```python
# importing matplotlib library to create hist() function to generate histograms for all numeric columns
import matplotlib.pyplot as plt
```

```python
# Select only the numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])
```
