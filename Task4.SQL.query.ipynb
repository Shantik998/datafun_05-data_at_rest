## 1-SELECT queries

```python
# get only the columns first and last from the authors table
pd.read_sql('SELECT first, last FROM authors',connection)
```

## 2-WHERE clause

```python
# select title, edition, and copyright for all books with copyright years greater than 2016
pd.read_sql("""SELECT title, edition, copyright
                FROM titles
                WHERE copyright > '2016'""", connection)
```

```python
# select all authors whose last name starts with D
pd.read_sql("""SELECT id, first, last
               FROM authors
               WHERE last LIKE 'D%'""", connection, index_col=['id'])
```

```python
# select all authors whose first name has b as the second letter
pd.read_sql("""SELECT id, first, last
               FROM authors
               WHERE first LIKE '_b%'""", connection, index_col=['id'])
```

## 3-ORDER BY clause

```python
# order by asceding order using ASC function
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection)
```

```python
# Sorting by multiple columns
# sort the authors name by last name
pd.read_sql("""SELECT id, first, last
               FROM authors
               ORDER BY last, first""", connection, index_col=['id'])
```

```python
# Sort authors by last name in descending order and first in ascending order
pd.read_sql("""SELECT id, first, last
               FROM authors
               ORDER BY last DESC, first ASC""",
            connection, index_col=['id'])
```

```python
# combining the WHERE and ORDER BY clauses 
# select  copy right from titles where title ends with 'How to Program', sort by title ascending
pd.read_sql("""SELECT isbn, title, edition, copyright
               FROM titles
               WHERE title LIKE '%How to Program'
               ORDER BY title""",
            connection)
```

## 4- INNER JOIN merging data from multiple tables

```python
#  merging first , last and ISBN from authors 
pd.read_sql("""SELECT first, last, isbn
               FROM authors
               INNER JOIN author_ISBN
                   ON authors.id = author_ISBN.id
               ORDER BY last, first""",
            connection)
```

## 5- INSERT INTO Statement

```python
#use of connection cursor method 
cursor = connection.cursor()
cursor = cursor.execute("""INSERT INTO authors (first, last)
                           VALUES ('Sue', 'Red')""")
```

```python
# query the authers table content
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col=['id'])
```

## 6- UPDATE statement

```python
# to modify existing vales and Update 'Sue Red' to 'Sue Black'
cursor = cursor.execute("""UPDATE authors SET last = 'Black'
                           WHERE last = 'Red' AND first = 'Sue'""")
```

```python
# Check number of records updated
cursor.rowcount
```

```python
# View authors table to confirm update by listing the authors table contents
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col=['id'])
```
